```python
import os
import pickle

import mediapipe as mp
import cv2
import matplotlib.pyplot as plt
```

```python
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
mp_drawing_styles = mp.solutions.drawing_styles
```

```python
hands = mp_hands.Hands()
```

```python
for img_path in os.listdir(os.path.join('./Training Set','A')):
  print(os.path.join('./Training Set','A',img_path))
```

```python
data = []
labels = []
Train_DIR = './Training Set'
for dir in os.listdir(Train_DIR):
  count = 0
  for img_path in os.listdir(os.path.join(Train_DIR,dir)):
    if count > 1500:
      break
    co_ordinates = []

    img = cv2.imread(os.path.join(Train_DIR,dir,img_path))
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    results = hands.process(img_rgb)
    if results.multi_hand_landmarks:
      for hand_landmarks in results.multi_hand_landmarks:
          for i in range(len(hand_landmarks.landmark)):
              x = hand_landmarks.landmark[i].x
              y = hand_landmarks.landmark[i].y

              co_ordinates.append(x)
              co_ordinates.append(y)
      data.append(co_ordinates)
      labels.append(i)
      count+=1
```
